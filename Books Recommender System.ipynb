### Popularity based Recommender System:

```python
ratings_with_bnames= ratings.merge(books, on='ISBN')
```

```python
ratings_with_bnames.head(3)
```

```python
num_of_ratings_df=ratings_with_bnames.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_of_ratings_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
num_of_ratings_df
```

```python
# Now we will calculate average ratings:
```

```python
avg_ratings_df= ratings_with_bnames.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_ratings_df.rename(columns={'Book-Rating':'avg_ratings'}, inplace=True)
avg_ratings_df
```

```python
popular_df= num_of_ratings_df.merge(avg_ratings_df, on='Book-Title')
popular_df
```

```python
popular_df[popular_df['num_ratings']>=250]
```

```python
popular_books= popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings', ascending=False).head(50)
popular_books
```

```python
books.head(1)
```

```python
popular_books= popular_books.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

### Collaborative Filtering Based Recommender System:

```python
a=ratings_with_bnames.groupby('User-ID').count()['Book-Rating']>200
user_rating_index= a[a].index
```

```python
filtered_rating= ratings_with_bnames[ratings_with_bnames['User-ID'].isin(user_rating_index)]
filtered_rating.head(3)
```

```python
b= filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books= b[b].index
famous_books[:3]
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings[:3]
```

```python
p_table=final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
# Fill all NaN values with 0.
```

```python
p_table.fillna(0, inplace=True)
```

```python
p_table
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score= cosine_similarity(p_table)
```

```python
similarity_score
```

```python
similarity_score.shape
```

```python
len(similarity_score[5])
```

```python
np.where(p_table.index=='4 Blondes')       # To fetch the index
```

```python
def book_recommend(book_name):
    index= np.where(p_table.index==book_name)[0][0]
    similar_items= sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1], reverse=True,)[1:5]
    
    data=[]
    for i in similar_items:
        
        item=[]
        temp_df= books[books['Book-Title']== p_table.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
    
        data.append(item)
    
    return data
```

```python
book_recommend('1984')
```

```python
book_recommend('Message in a Bottle')
```
